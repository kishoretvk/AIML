### Analysis of spread and distribution of every feature in the dataset

```python
# Visualize data for each feature. Comparing data distribution of Products for each feature 
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

sns.histplot(data=data, x="Age", hue="Product", multiple="dodge", shrink=.8, bins=5, kde=True, ax=axs[0, 0])
sns.histplot(data=data, x="Education", hue="Product", multiple="dodge", shrink=.8, bins=5, kde=True, ax=axs[0, 1])
sns.histplot(data=data, x="Income", hue="Product", multiple="dodge", shrink=.8, bins=4, kde=True, ax=axs[0, 2])
sns.histplot(data=data, x="Usage", hue="Product", multiple="dodge", shrink=.8, kde=True, ax=axs[1, 0])
sns.histplot(data=data, x="Fitness", hue="Product", multiple="dodge", shrink=.8, kde=True, ax=axs[1, 1])
sns.histplot(data=data, x="Miles", hue="Product", multiple="dodge", shrink=.8, bins=5, kde=True, ax=axs[1, 2])
sns.histplot(data=data, x="Gender", hue="Product", multiple="dodge", shrink=.8, ax=axs[2, 0])
sns.histplot(data=data, x="MaritalStatus", hue="Product", multiple="dodge", shrink=.8, ax=axs[2, 1])
sns.histplot(data=data, x="Product", shrink=.8, ax=axs[2, 2])

plt.show()
```

#### Analysis of each feature

1. **Age**: <40 Years customers are buying more products than other customers.
2. **Education**: <18 years are buying TM195 and TM498. >=16 years customers like TM798.
3. **Income**: <=60000 income customers like TM195 and TM498. >=60000 income customers like TM798.
4. **Usage**: <=4 times a week are using TM195 and TM498. >=4 times a week are using TM798.
5. **Fitness**: <=4 rated fitness score customers are using TM195 and TM498. >=4 rated fitness score customers are using TM798.
6. **Miles**: <200 miles customers are using TM195 and TM498. >=100 miles customers are using TM798.
7. **Gender**: Male and Female like TM195 more. But Female don't like TM798.
8. **MaritalStatus**: Partnered customers using more products than single customers. Partnered and Single both likes TM195.
9. **Product**: Customers more likes TM195 than TM498 and than TM798.

### Analysis of interaction between features, in the dataset

```python
sns.heatmap(data.corr(), annot=True) # checking the correlation of the different variables.
```

```python
# using pairplot to understand the correlation of all the numerical columns. 
sns.set_theme(style="ticks")
sns.pairplot(data=data, hue='Product')
```

**Education and Income are correlated variables. Where Usage, Fitness and Miles are correlated variables. Ploting the graphs using Income, Usage and Age as numerical variables.**

```python
fig, axs = plt.subplots(2,3, figsize=(15, 15))
sns.boxplot(x='Gender', y='Income', hue='Product', data=data, ax=axs[0,0])
sns.boxplot(x='Gender', y='Usage', hue='Product', data=data, ax=axs[0,1])
sns.boxplot(x='Gender', y='Age', hue='Product', data=data, ax=axs[0,2])
sns.boxplot(x='MaritalStatus', y='Income', hue='Product', data=data, ax=axs[1,0])
sns.boxplot(x='MaritalStatus', y='Usage', hue='Product', data=data, ax=axs[1,1])
sns.boxplot(x='MaritalStatus', y='Age', hue='Product', data=data, ax=axs[1,2])
plt.show()
```
